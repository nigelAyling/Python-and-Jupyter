```python
#Create a new unshuffled deck. Use only numbers 1 to 52.

import random

def create_deck():
    new_deck = []
    for card in range (1,53):
        #print (card)
        new_deck.append(card)
    return new_deck

def shuffle_deck(deck):
    cards_left = 52
    shuffled_deck = []
    while cards_left > 0:
        n = random.randint(0,cards_left)
        #print ("this is the random number- ",n)
        card = deck[n-1]
        del deck[n-1]
        shuffled_deck.append(card)
        cards_left = cards_left - 1
        #print ("New pack- ")
        #print (deck)
        #print ("shuffled pack")
        #print (shuffled_deck)
        #print()
    return shuffled_deck

def display_card(card_num):
    face = ""
    if card_num < 14:
        suit = "of Clubs"
        value = card_num
    elif card_num <27:
        suit = "of Diamonds"
        value = card_num - 13
    elif card_num <40:
        suit = "of Hearts"
        value = card_num - 26
    elif card_num <53:
        suit = "of Spades"
        value = card_num - 39
    if value == 1:
        face = "Ace "
    elif value == 11:
        face = "Jack"
    elif value == 12:
        face = "Queen "
    elif value == 13:
        face = "King "
    else:
        face = str(value)
    card_value = face +" " +suit
    return card_value
    
        
```

```python

pack = create_deck()
cards_to_deal = shuffle_deck(pack)
print(cards_to_deal)
print("")
print("")
for n in cards_to_deal:
    print(display_card(n))
    #print (new_card)
    

```
